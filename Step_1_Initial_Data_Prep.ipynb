```python
import pandas as pd
```

# Initial Data loading

This section of the code extracts county 1001 from the larger train, test and validation csv file from Kaggle’s ‘Predict Droughts using Weather & Soil Data’ dataset. The data can be downloaded from this link: https://www.kaggle.com/cdminix/us-drought-meteorological-data. This data for a single county is appended to a single dataframe containing all time data present and saved to a new csv file for ease of use

```python
def load(data, county):
    """Load in data from all counties and extract single county data"""
    county = data.xs(county)
    county.sort_index(inplace=True)
    del data
    return county
```

### Load test data

```python
test = pd.read_csv('test_timeseries/test_timeseries.csv').set_index(['fips', 'date'])
```

```python
test_data = load(test, 1001)
del test
```

```python
test_data
```

### Load training data

```python
train = pd.read_csv('train_timeseries/train_timeseries.csv').set_index(['fips', 'date'])
```

```python
train_data = load(train, 1001)
del train
```

```python
train_data
```

### Load validation data

```python
val = pd.read_csv('validation_timeseries/validation_timeseries.csv').set_index(['fips', 'date'])
```

```python
val_data = load(val, 1001)
del val
```

```python
val_data
```

### Append test, train and validation data for county 1001 and save to csv

```python
full_data = train_data.append(val_data)
full_data = full_data.append(test_data)
full_data.to_csv('full_1001.csv')
```
